```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('/content/house_prices.csv')

heatmap_data = data.pivot_table(index='state', columns='date', values='house_price_index')

heatmap_data.columns = pd.to_datetime(heatmap_data.columns)

plt.figure(figsize=(16, 10))
plt.imshow(heatmap_data, aspect='auto', cmap='viridis', interpolation='nearest')
plt.colorbar(label='House Price Index')
plt.title('House Price Index Across States Over Time')
plt.xlabel('Date')
plt.ylabel('State')

plt.xticks(ticks=np.arange(0, len(heatmap_data.columns), step=12),
           labels=heatmap_data.columns[::12].strftime('%Y-%m'))
plt.yticks(ticks=np.arange(len(heatmap_data.index)), labels=heatmap_data.index)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('/content/house_prices.csv')

michigan_data = data[data['state'] == 'Michigan']

plt.figure(figsize=(12, 6))
plt.plot(michigan_data['date'], michigan_data['house_price_index'], marker='o', linestyle='-', color='b')
plt.title('House Price Index Over Time in Michigan')
plt.xlabel('Date')
plt.ylabel('House Price Index')
plt.grid(True)

plt.xticks(pd.date_range(start=michigan_data['date'].min(), end=michigan_data['date'].max(), freq='6M'),
           labels=pd.date_range(start=michigan_data['date'].min(), end=michigan_data['date'].max(), freq='6M').strftime('%Y-%m'),
           rotation=45)

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

housing_data = pd.read_csv('/content/Housing.csv')

average_price_mainroad = housing_data.groupby('mainroad')['price'].mean()

plt.figure(figsize=(8, 6))
average_price_mainroad.plot(kind='line', marker='o')
plt.title('Average House Price vs. Proximity to Main Road')
plt.xlabel('Main Road')
plt.ylabel('Average Price')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.grid(True)
plt.show()
```
